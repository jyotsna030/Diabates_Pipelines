```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical variable (if `Sex` is present)
if 'Sex' in df.columns:
    label_encoder = LabelEncoder()
    df['Sex'] = label_encoder.fit_transform(df['Sex'])

# Example: Group age into categories (optional, based on dataset)
age_bins = [0, 18, 35, 50, 65, 100]
age_labels = ['Child', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Select numeric columns for scaling
numeric_features = ['BMI', 'Age', 'MentHlth', 'PhysHlth']  # Example columns
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])

print(df[numeric_features].head())
```

```python
# Detect outliers using IQR
for col in numeric_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print(df[numeric_features].describe())
```

```python
# Exclude non-numeric columns from correlation matrix calculation
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()

# Select correlated features based on threshold
correlated_features = correlation_matrix['Diabetes_012'].abs().sort_values(ascending=False)
selected_features = correlated_features[correlated_features > 0.1].index.tolist()
selected_features.remove('Diabetes_012')  # Exclude target column

print("Selected Features:", selected_features)
```

```python
# Split data into X (features) and y (target)
X = df[selected_features]
y = df['Diabetes_012']

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
# Save processed data to CSV
X.to_csv('processed_features.csv', index=False)
y.to_csv('processed_target.csv', index=False)
```

```python
def preprocess_data(df):
    # Feature Engineering
    if 'Sex' in df.columns:
        df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

    # Encode Age_Group to numeric if it exists
    if 'Age_Group' in df.columns:
        df['Age_Group'] = LabelEncoder().fit_transform(df['Age_Group'].astype(str))

    # Handle missing values: Fill with median (numeric) or mode (categorical)
    for col in df.select_dtypes(include='number').columns:
        if df[col].isnull().sum() > 0:
            df[col] = df[col].fillna(df[col].median())



    # Scale numeric features
    numeric_features = ['BMI', 'Age', 'MentHlth', 'PhysHlth']  # Add other numeric columns as needed
    scaler = StandardScaler()
    df[numeric_features] = scaler.fit_transform(df[numeric_features])

    # Handle outliers
    for col in numeric_features:
        Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound, upper_bound = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

    # Select features
    correlation_matrix = df.corr()
    correlated_features = correlation_matrix['Diabetes_012'].abs().sort_values(ascending=False)
    selected_features = correlated_features[correlated_features > 0.1].index.tolist()
    selected_features.remove('Diabetes_012')  # Exclude target column

    X = df[selected_features]
    y = df['Diabetes_012']

    # Final check for missing values in X
    if X.isnull().sum().any():
        print("Filling remaining missing values in features...")
        X.fillna(X.median(), inplace=True)

    return X, y

```

```python
# Test the preprocess_data function
X, y = preprocess_data(df)

# Validate shapes and missing values
assert not X.isnull().any().any(), "Missing values in features"
assert not y.isnull().any(), "Missing values in target"
print("Data Preprocessing Passed!")

print("Features Sample:")
print(X.head())
```
